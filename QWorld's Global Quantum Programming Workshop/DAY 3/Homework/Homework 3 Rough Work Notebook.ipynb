```python
from qiskit import *
```

```python
q2=QuantumRegister(3,"qreg")
c2=ClassicalRegister(3,"creg")
qc2=QuantumCircuit(q2,c2)

qc2.x(q2[1])
qc2.barrier()

qc2.x(q2[2])
qc2.ccx(q2[2],q2[1],q2[0])
qc2.x(q2[2])

qc2.barrier()

qc2.measure(q2,c2)
job=execute(qc2,Aer.get_backend('qasm_simulator'),shots=100)
counts=job.result().get_counts(qc2)
print(counts)

qc2.draw('mpl')
```
